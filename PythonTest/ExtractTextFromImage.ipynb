```python
from PIL import Image

image_path = 'Google.jpg'

Image.open(image_path)
```

```python
from PIL import Image
import pytesseract

# En Windows, especifica la ruta completa de Tesseract si no est√° en PATH
pytesseract.pytesseract.tesseract_cmd = r'C:\Program Files\Tesseract-OCR\tesseract.exe'


# Open on image file

image = Image.open(image_path)
extracted_text = pytesseract.image_to_string(image)
print(extracted_text)
```

```python
# Obtener el resultado en formato HOCR (HTML)
hocr_content = pytesseract.image_to_pdf_or_hocr(image, extension='hocr')

# Guardar el contenido HOCR en un archivo HTML
with open("output.html", "wb") as f:
    f.write(hocr_content)

print("Archivo HTML guardado como output.html")
```
