```python
import PyPDF2
import pyttsx3
```

```python
# prompt
# pdf_filename = input("Enter de pdf file").strip()
pdf_filename = "app.pdf"
```

```python
import re
try:
    with open(pdf_filename, "rb") as pdf_file:
        pdf_reader = PyPDF2.PdfReader(pdf_file)

        speak = pyttsx3.init()

        for page_num in range(len(pdf_reader.pages)):
            page = pdf_reader.pages[page_num]
            text =  page.extract_text()
            if text:
                #speak.say(text)
                text.replace("\n", " ")
                text.strip()
                text.replace("\n", " ")
                cleaned_text = re.sub(r'\s+', ' ', text).strip()

                print(cleaned_text)
                speak.save_to_file(cleaned_text, pdf_filename +  "_page_" + str(page_num) + ".mp3")
                speak.runAndWait()

        speak.stop()

        print("Audiobook creation completed.")
except FileNotFoundError:
    print("")
except Exception as e:
    print("error")
```
